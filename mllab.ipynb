```python
import json, nltk

def clear_not_alpha_tokens(words):
    return [token.lower() for token in words if token.isalpha()]

def clear_stop_words(tokens):
    stop_words = nltk.corpus.stopwords.words()
    return [t for t in tokens if (t not in stop_words)]

def get_most_freq_words(sentence):
    tokens = nltk.word_tokenize(sentence)
    tokens = clear_not_alpha_tokens(tokens)
    tokens = clear_stop_words(tokens)
    freq_words = nltk.FreqDist(tokens)
    return freq_words.most_common()

f = open('dataset.json').read()
data = json.loads(f)
sentences = '\n'.join([s['text'].lower() for s in data])
res = get_most_freq_words(sentences)
print(res)
```
